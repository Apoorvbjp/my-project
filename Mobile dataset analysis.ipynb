```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mob =  pd.read_csv('mobile.csv') # read csv file
```

```python
mob.isnull().sum() # check there is any null por not
```

```python
mob['blue'].value_counts() #check the count of bluetooth 
```

```python
mob['blue'].fillna(0, inplace=True) #full fill the null value in bluetooth
```

```python
mob.isnull().sum()
```

```python
mob['dual_sim'].value_counts()
```

```python
mob['dual_sim'].fillna(1, inplace=True)
```

```python
mob.isnull().sum()
```

```python
mob['mobile_wt'].fillna(np.mean(mob['mobile_wt']),inplace=True)
```

```python
mob.isnull().sum()
```

```python
mob.duplicated().sum()# check the duplicate value in dataset
```

```python
count_dual_sim = mob['dual_sim'].value_counts()
count_dual_sim
```

```python
#create a bar chart to see how much mobile are supporting the dual sim 
plt.bar(count_dual_sim.index,count_dual_sim.values)
plt.xlabel('dual sim or not')
plt.ylabel('Number of mobiles')
plt.show()
```

```python
#create a pie chart to see how much mobile are supporting the 4G
plt.pie(count_dual_sim.values,labels = count_dual_sim.index , shadow = True , explode = (0,0.1), autopct ='%1.1f%%')
plt.legend()
plt.show()
```

```python
#create bar chart to see how much mobile support the touch screen
plt.bar(mob['touch_screen'].value_counts().index,mob['touch_screen'].value_counts().values)
plt.xlabel('Touch Screen Support')
plt.ylabel('Count of Mobile')
plt.show()
```
